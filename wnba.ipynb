```python
import pandas as pd

draftdf = pd.read_csv('wnbadraft.csv')
draftdf.head()
```

## Cleaning

```python
# check type
draftdf.dtypes
```

```python
# check NaNs
draftdf.isnull().sum()

# replace nans in former
draftdf['former'] = draftdf['former'].fillna('US college')

# replace nans in college
draftdf['college'] = draftdf['college'].fillna('International')

# exploring rows with missing players
draftdf[draftdf['player'].isna()]
draftdf = draftdf.dropna(subset=['player']) # dropping 2 rows with missing values

# replace nans with 0
draftdf[['games', 'win_shares', 'win_shares_40', 'points', 'total_rebounds', 'minutes_played', 'assists']] = draftdf[['games', 'win_shares', 'win_shares_40', 'points', 'total_rebounds', 'minutes_played', 'assists']].fillna(0)

draftdf.isnull().sum()

```

## Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# get top 5 draft picks
top5 = draftdf[draftdf['overall_pick'] <= 5]
team_top_picks = top5['team'].value_counts().reset_index()
team_top_picks.columns = ['team', 'top_5_count']

team_top_picks = team_top_picks.sort_values(by='top_5_count', ascending=True)

colors = cm.get_cmap('Oranges', len(team_top_picks))

plt.figure(figsize=(10, 12))
plt.barh(team_top_picks['team'], team_top_picks['top_5_count'], color=colors(range(len(team_top_picks))))
plt.title('Number of Top 5 Draft Picks by Team (1997-2022)', fontweight='bold')
plt.xlabel('Top 5 Draft Picks', labelpad = 15)
plt.ylabel('Team', labelpad=20)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your draft data (assumed to be already done)
# draftdf = pd.read_csv('path_to_your_data.csv')

# Define the teams of interest
teams_of_interest = ['Chicago Sky', 'Minnesota Lynx', 'Washington Mystics', 'Dallas Wings', 'Indiana Fever']

# Filter the dataset for top 5 picks from these teams
top5_picks = draftdf[(draftdf['overall_pick'] <= 5) & (draftdf['team'].isin(teams_of_interest))]

# Calculate total points and win shares for these teams
performance_summary = top5_picks.groupby('team').agg({
    'points': 'sum',
    'win_shares': 'sum',
    'games': 'sum'
}).reset_index()

# Calculate average points per game and average win shares per game
performance_summary['avg_points_per_game'] = performance_summary['points'] / performance_summary['games']
performance_summary['avg_win_shares_per_game'] = performance_summary['win_shares'] / performance_summary['games']

# Visualize the total points and win shares for these teams
plt.figure(figsize=(12, 6))

# Bar plot for total points
plt.subplot(1, 2, 1)
plt.bar(performance_summary['team'], performance_summary['points'], color='cornflowerblue')
plt.title('Total Points from Top 5 Picks (1997-2022)', fontweight='bold')
plt.xlabel('Team')
plt.ylabel('Total Points')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Bar plot for total win shares
plt.subplot(1, 2, 2)
plt.bar(performance_summary['team'], performance_summary['win_shares'], color='orange')
plt.title('Total Win Shares from Top 5 Picks (1997-2022)', fontweight='bold')
plt.xlabel('Team')
plt.ylabel('Total Win Shares')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
